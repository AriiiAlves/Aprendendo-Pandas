# Tipos de variveis

## Tipos de dados

- **Numrico (quantitativo)** : Dados que podem ser expressados em uma escala numrica
    - **Discreto**: Dados que, no geral, podem ter apenas nmeros **inteiros**, geralmente associados  contagem (nmero de quartos em um hotel, nmero de nascimentos deste ano, idade (em anos) de uma pessoa)
    - **Contnuo**: Dados que podem ter qualquer valor em um intervalo contnuo (**float** ou **double**). Geralmente associado  medio (altura; peso; rea; tempo).
<br><br>
- **Categrico (qualitativo)**: Dados que podem ter apenas um conjunto de valores que representam categorias/grupos
    - **Nominal**: No h uma ordem entre as categorias (cor do cabelo; idioma; gnero literrio; aprovado? (sim, no)).
    - **Ordinal**: H uma ordem entre as categorias (tamanho de roupa (P, M, G); Escala de satisfao (ruim, mdio, bom); Meses do ano)
    - **Identificador**: Possui apenas um **nico indivduo** por categoria (ID, RG...).
<br><br>
- **Outros tipos**
    - **Texto**: String
    - **Datas**: String que representa datas. Pode ser convertida em novas variveis, como meses, anos...



## Casos especiais

### Variveis discretas tratadas como variveis contnuas

O dinheiro ou preo de algo varia em passos de 1 centavo, ento  uma varivel discreta.

Porm, se voc est lidando com centenas de reais, os passos so to pequenos que ele pode ser tratado como uma varivel contnua.

### Variveis categricas representadas com nmeros

Tais nmeros no tm sentido "numrico", no sentido de que no faz sentido utilizar tais nmeros em operaes como soma, subtrao, etc.

- Escala de satisfao de um atendimento
    - 0: Pssimo
    - 1: Ruim
    - 2: Bom
    - 3: Excelente
- Escolaridade
    - 1 (grau)
    - 2 (grau)
    - 3 (grau)
- Avaliao de filmes
    - Um filme nota 4.0 pode no se, necessariamente, 2 vezes melhor do que um filme nota 2.0
