# Tipos de variáveis

## Tipos de dados

- **Numérico (quantitativo)** : Dados que podem ser expressados em uma escala numérica
    - **Discreto**: Dados que, no geral, podem ter apenas números **inteiros**, geralmente associados à contagem (número de quartos em um hotel, número de nascimentos deste ano, idade (em anos) de uma pessoa)
    - **Contínuo**: Dados que podem ter qualquer valor em um intervalo contínuo (**float** ou **double**). Geralmente associado à medição (altura; peso; área; tempo).
<br><br>
- **Categórico (qualitativo)**: Dados que podem ter apenas um conjunto de valores que representam categorias/grupos
    - **Nominal**: Não há uma ordem entre as categorias (cor do cabelo; idioma; gênero literário; aprovado? (sim, não)).
    - **Ordinal**: Há uma ordem entre as categorias (tamanho de roupa (P, M, G); Escala de satisfação (ruim, médio, bom); Meses do ano)
    - **Identificador**: Possui apenas um **único indivíduo** por categoria (ID, RG...).
<br><br>
- **Outros tipos**
    - **Texto**: String
    - **Datas**: String que representa datas. Pode ser convertida em novas variáveis, como meses, anos...



## Casos especiais

### Variáveis discretas tratadas como variáveis contínuas

O dinheiro ou preço de algo varia em passos de 1 centavo, então é uma variável discreta.

Porém, se você está lidando com centenas de reais, os passos são tão pequenos que ele pode ser tratado como uma variável contínua.

### Variáveis categóricas representadas com números

Tais números não têm sentido "numérico", no sentido de que não faz sentido utilizar tais números em operações como soma, subtração, etc.

- Escala de satisfação de um atendimento
    - 0: Péssimo
    - 1: Ruim
    - 2: Bom
    - 3: Excelente
- Escolaridade
    - 1 (grau)
    - 2 (grau)
    - 3 (grau)
- Avaliação de filmes
    - Um filme nota 4.0 pode não se, necessariamente, 2 vezes melhor do que um filme nota 2.0
